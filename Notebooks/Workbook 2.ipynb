```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('../Data/Speed Dating Data.csv', encoding = "ISO-8859-1", engine='python')
pd.set_option("display.max_columns", None)
data
```

```python
# def partner_given_rating(data):
self = ['attr3_1','intel3_1','sinc3_1','fun3_1','amb3_1']
market = ['attr5_1','intel5_1','sinc5_1','fun5_1','amb5_1']
self_ratings = data[['attr3_1','intel3_1','sinc3_1','fun3_1','amb3_1','attr5_1','intel5_1','sinc5_1','fun5_1','amb5_1']].copy()
lst = ['attr_o','sinc_o','intel_o','fun_o','amb_o','shar_o']
copy_data = data.iloc[:,:33].fillna(0)
newdf = pd.concat([copy_data,self_ratings], axis = 1).fillna(0)
newdf['Given Partner Rating'] = (newdf[lst].sum(axis = 1)).fillna(0)
newdf['iid Self Given Rating'] = (newdf[self].sum(axis = 1)).fillna(0)
newdf['Total Self Market Rating'] = (newdf[market].sum(axis = 1)).fillna(0)

newdf.loc[(copy_data['iid'] == 243)]
```

```python
df1 = newdf.groupby('pid')
```

```python
lookup = pd.DataFrame(newdf[['iid','iid Self Given Rating','Total Self Market Rating']]).drop_duplicates().rename(columns={'iid':'pid'}).set_index('pid')
lookup
```

```python
focus = ['like_o','prob_o','Given Partner Rating']
newdf2 = newdf.groupby(['pid','iid'])[focus].agg('mean').round(2).fillna(0)
averages = newdf2.join(lookup, on = 'pid').rename(columns={"iid Self Given Rating": "Total Self Given Rating"})
averages['% Error Partner Rating vs Self'] = (((averages['Total Self Given Rating'] - averages['Given Partner Rating'])/ averages['Given Partner Rating']).round(2)*100).fillna(0)
averages['% Error Market Rating vs Self'] = (((averages['Total Self Market Rating'] - averages['Given Partner Rating'])/ averages['Given Partner Rating']).round(2)*100).fillna(0)

averages
```
