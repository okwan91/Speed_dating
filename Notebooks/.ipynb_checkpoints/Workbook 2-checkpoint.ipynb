```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('../Data/Speed Dating Data.csv', encoding = "ISO-8859-1", engine='python')
pd.set_option("display.max_columns", None)
data
```

# Organizing and adding new columns:

I've identified the columns I want to use to test my hypothesis and also create new columns that contain the sum of participants scores that were given by their partner or given to themselves. 

```python
# def partner_given_rating(data):
self = ['attr3_1','intel3_1','sinc3_1','fun3_1','amb3_1']
market = ['attr5_1','intel5_1','sinc5_1','fun5_1','amb5_1']
self_ratings = data[['attr3_1','intel3_1','sinc3_1','fun3_1','amb3_1','attr5_1','intel5_1','sinc5_1','fun5_1','amb5_1']].copy()
lst = ['attr_o','sinc_o','intel_o','fun_o','amb_o','shar_o']
copy_data = data.iloc[:,:33].fillna(0)
newdf = pd.concat([copy_data,self_ratings], axis = 1).fillna(0)
newdf['Given Partner Rating'] = (newdf[lst].sum(axis = 1)).fillna(0)
newdf['iid Self Given Rating'] = (newdf[self].sum(axis = 1)).fillna(0)
newdf['Total Self Market Rating'] = (newdf[market].sum(axis = 1)).fillna(0)

newdf.loc[(copy_data['iid'] == 243)]
```

# Lookup table of people's ratings for themselves

```python
lookup = pd.DataFrame(newdf[['iid','iid Self Given Rating','Total Self Market Rating']]).drop_duplicates().rename(columns={'iid':'pid'}).set_index('pid')
lookup
```

# Focused Data:
Creating a new grouped table with focus columns and joining it with our lookup table that we made before so that we are able to associate each participants scores given by their partner.
    

```python
focus = ['like_o','prob_o','Given Partner Rating']
newdf2 = newdf.groupby(['pid','iid'])[focus].agg('mean').round(2).fillna(0)
averages = newdf2.join(lookup, on = 'pid').rename(columns={"iid Self Given Rating": "Total Self Given Rating"})
averages['% Error Partner Rating vs Self'] = ((averages['Given Partner Rating'] - averages['Total Self Given Rating'])/ averages['Given Partner Rating']).abs().replace(np.inf,0).round(2)*100
averages['% Error Market Rating vs Self'] = ((averages['Given Partner Rating'] - averages['Total Self Market Rating'])/ averages['Given Partner Rating']).abs().replace(np.inf,0).round(2)*100

averages
```

# People who gave themselves a market rating:
While every participant gave themselves a rating of themselves in general, they may not have given a rating for themselves in the market. Below table isolates the people who did give themselves a market rating.

```python
notnull_ = averages.loc[(averages.loc[:,'Total Self Market Rating'] != 0.0)]
notnull_avg = notnull_.loc[(notnull_.loc[:,'% Error Market Rating vs Self'] != 0.0)]

notnull_avg
```

# How accurate are people?

## People who gave themselves a market rating:

```python
print(f'People who gave themselves a market rating on average had a {notnull_avg.iloc[:,-1].mean()}% of error.')
```

## People who only gave self rated themselves:

```python
print(f'How people\'s ratings of themselves on average had a {averages.iloc[:,-2].mean()}% of error.')
```
